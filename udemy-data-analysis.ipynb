# 1. Dataset Description

The dataset contains information about Udemy courses, including details such as course titles, instructors, ratings, prices, number of reviews, and content duration. It provides insights into course popularity, pricing strategies, and engagement levels.

## Columns:
 - course_id: Unique identifier for each course.
 - course_title: Name of the course.
 - url: Link to the course on Udemy.
 - is_paid: Whether the course is free or paid.
 - price: Cost of the course.
 - num_subscribers: Number of students enrolled.
 - num_reviews: Total reviews given.
 - num_lectures: Total lectures in the course.
 - num_content_hours: Course duration in hours.
 - level: Course difficulty (Beginner, Intermediate, etc.).
 - published_date: Date when the course was released.
 - category: Course topic category (e.g., Business, Development).

#### 1. Price column supposed to be **int** not **object**, so let's check it's value 

```python
df_udemy['price'].unique()
```

```python
# Replacing 'Free' with 0
df_udemy['price'] = df_udemy['price'].replace('Free', 0)

# Converting the 'price' column to integer type
df_udemy['price'] = df_udemy['price'].astype(int)

print(df_udemy['price'].unique())
```

#### 2. content_duration is an **object** type, so converting it to a **timedelta** type would make it more useful for analysis

```python
def extract_duration(duration):
    if 'hour' in duration or 'hours' in duration:
        # Extract the numeric part for hours
        hour_part = duration.split('hour')[0].strip()
        if 'hours' in duration:
            hour_part = duration.split('hours')[0].strip()
        if hour_part.replace('.', '', 1).isdigit():  # Check if it's a valid number (including floats)
            return float(hour_part), 'hours'
    
    if 'min' in duration or 'mins' in duration:
        # Extract the numeric part for minutes
        minute_part = duration.split('min')[0].strip()
        if 'mins' in duration:
            minute_part = duration.split('mins')[0].strip()
        if minute_part.replace('.', '', 1).isdigit():  # Check if it's a valid number (including floats)
            return float(minute_part), 'mins'
    
    return None, None  # Return None for invalid formats
```

```python
# Apply the function and create new columns for numerical value and unit
df_udemy[['duration_value', 'duration_unit']] = df_udemy['content_duration'].apply(
    lambda x: pd.Series(extract_duration(x))
)

# Drop the original 'content_duration' column
df_udemy.drop(columns=['content_duration'], inplace=True)
```

```python
df_udemy[['duration_value', 'duration_unit']]
```

#### 3. published_timestamp column is currently stored as an **object** which is not suitable for a column that represents a **date**

```python
# Convert the 'published_timestamp' column to datetime
df_udemy['published_timestamp'] = pd.to_datetime(df_udemy['published_timestamp'], errors='coerce')

df_udemy['published_timestamp']
```

# 4. Visualization and Analytical Questions

### 1. What is the distribution of course prices?

```python
sns.histplot(df_udemy['price'], kde=True, bins=30)
plt.title('Distribution of Course Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

### 2. How does the number of subscribers vary by course subject?

```python
sns.boxplot(x='subject', y='num_subscribers', data=df_udemy)
plt.title('Number of Subscribers by Subject')
plt.xlabel('Subject')
plt.ylabel('Number of Subscribers')
plt.xticks(rotation=45)
plt.show()
```

### 3. What is the relationship between course duration and the number of subscribers?

```python
sns.scatterplot(x='duration_value', y='num_subscribers', data=df_udemy)
plt.title('Course Duration vs Number of Subscribers')
plt.xlabel('Duration (hours)')
plt.ylabel('Number of Subscribers')
plt.show()
```
